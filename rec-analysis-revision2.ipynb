```python
from rdkit import Chem

# Function to check if a molecule contains 3, 4, 5, 7, or 9 carbon rings
def find_specific_carbon_rings(molecule):
    # Get ring information
    ring_info = molecule.GetRingInfo()

    # Initialize a dictionary to count the number of specific carbon rings
    ring_sizes = {3: 0, 4: 0, 5: 0, 7: 0, 8: 0, 6: 0}

    # Initialize lists for 5-membered and 6-membered rings
    five_membered_rings = []
    six_membered_rings = []

    # Check if the molecule contains any ring structure
    if ring_info.NumRings() > 0:
        # Loop through each ring and check the number of carbon atoms in the ring
        for ring in ring_info.AtomRings():
            # Count how many carbon atoms are in this ring
            carbon_count = sum(1 for atom_idx in ring if molecule.GetAtomWithIdx(atom_idx).GetAtomicNum() == 6)

            # If the carbon count is one of the desired sizes (3, 4, 5, 7, or 9), increment the counter
            if carbon_count in ring_sizes:
                ring_sizes[carbon_count] += 1
                # Add the ring to the respective list for 5-membered or 6-membered rings
                if carbon_count == 5:
                    five_membered_rings.append(ring)
                elif carbon_count == 6:
                    six_membered_rings.append(ring)

    return ring_sizes, five_membered_rings, six_membered_rings

# List of SMILES strings (example including rings with 3, 4, 5, 7, and 9 carbon atoms)
smiles_list = [
    'C1CC1',           # Cyclopropane (3 carbon ring)
    'C1CCC1',          # Cyclobutane (4 carbon ring)
    'C1CCCC1',         # Cyclopentane (5 carbon ring)
    'C1CCCCCC1',       # Cyclohexane (6 carbon ring)
    'C1CCCCC1',        # Cycloheptane (7 carbon ring)
    'C1CCCCCCCC1',     # Cyclooctane (8 carbon ring)
    'C1CCCCCCCCC1',    # Cyclononane (9 carbon ring)
    'C1OCCCC1',        # Furan (5-membered ring with oxygen)
    'C1CCNCC1'         # Pyrrolidine (5-membered ring with nitrogen)
]

# Convert SMILES to RDKit Molecule objects
molecules = [Chem.MolFromSmiles(smiles) for smiles in fullrecsmiles]

# Initialize separate lists for 5- and 6-membered rings
five_membered_ring_molecules = []
six_membered_ring_molecules = []

# Count how many molecules have specific carbon rings
ring_count = {3: 0, 4: 0, 5: 0, 7: 0, 8: 0, 6: 0}

for mol in molecules:
    if mol:
        ring_sizes, five_rings, six_rings = find_specific_carbon_rings(mol)
        # Add the counts of rings from this molecule to the overall counts
        for size in ring_count:
            ring_count[size] += ring_sizes[size]
        # Add the molecules with 5- and 6-membered rings to their respective lists
        if five_rings:
            five_membered_ring_molecules.append(mol)
        if six_rings:
            six_membered_ring_molecules.append(mol)

# Output the result
for size in ring_count:
    print(f"Number of {size}-carbon rings: {ring_count[size]}")

# Output the lists of molecules with 5- and 6-membered rings
print(f"\nMolecules with 5-membered carbon rings: {len(five_membered_ring_molecules)}")
for mol in five_membered_ring_molecules:
    print(Chem.MolToSmiles(mol))

print(f"\nMolecules with 6-membered carbon rings: {len(six_membered_ring_molecules)}")
for mol in six_membered_ring_molecules:
    print(Chem.MolToSmiles(mol))
```

```python
from rdkit import Chem

def calculate_degree_of_saturation(molecule):
    # Get all bonds in the molecule
    bonds = molecule.GetBonds()
    num_bonds = len(bonds)
    
    # Count hydrogens (the number of explicit hydrogens is the total number of hydrogens)
    num_hydrogens = sum(atom.GetTotalNumHs() for atom in molecule.GetAtoms())
    
    # Degree of saturation formula
    degree_of_saturation = (num_bonds - num_hydrogens) / num_bonds if num_bonds > 0 else 0.0
    return degree_of_saturation

def calculate_average_degree_of_saturation(smiles_list):
    total_degree_of_saturation = 0.0
    valid_molecules = 0
    
    for smiles in smiles_list:
        molecule = Chem.MolFromSmiles(smiles)
        if molecule:
            total_degree_of_saturation += calculate_degree_of_saturation(molecule)
            valid_molecules += 1
    
    if valid_molecules > 0:
        average_degree_of_saturation = total_degree_of_saturation / valid_molecules
    else:
        average_degree_of_saturation = 0.0
        
    return average_degree_of_saturation

# Example list of SMILES strings
smiles_list = [
    "CCO",          # Ethanol
    "CC(=O)O",      # Acetic acid
    "C1CCCCC1",     # Cyclohexane
    "C=CC=C",       # Propene
    "C#CC#C"        # Butyne
]

# Calculate the average degree of saturation
avg_degree = calculate_average_degree_of_saturation(fullrecsmiles)
print(f"Average degree of saturation: {avg_degree:.3f}")
```

```python
from rdkit import Chem
from rdkit.Chem import rdmolops

def check_ring_size(molecule):
    # Get the ring info for the molecule
    ring_info = molecule.GetRingInfo()
    ring_sizes = [len(ring) for ring in ring_info.AtomRings()]
    
    # Count the 5-membered and 6-membered rings
    count_5 = ring_sizes.count(5)
    count_6 = ring_sizes.count(6)
    
    # Return counts of 5-membered and 6-membered rings
    return count_5, count_6

def filter_molecules(smiles_list):
    valid_5_membered = []
    valid_6_membered = []
    
    for smiles in smiles_list:
        # Convert the SMILES string to a molecule object
        molecule = Chem.MolFromSmiles(smiles)
        
        if molecule is None:
            continue
        
        count_5, count_6 = check_ring_size(molecule)
        
        # Check if molecule has exactly one 5-membered or 6-membered ring
        if count_5 == 1 and count_6 == 0:  # Only a single 5-membered ring
            valid_5_membered.append(smiles)
        elif count_5 == 0 and count_6 == 1:  # Only a single 6-membered ring
            valid_6_membered.append(smiles)
    
    return valid_5_membered, valid_6_membered

# Example SMILES list
smiles_list = [
    "C1CCCC1",  # Cyclohexane (6-membered)
    "C1CCC1",   # Cyclopentane (5-membered)
    "C1CCCCC1", # Cyclohexane (6-membered)
    "C1CCCCO1", # Oxirane (5-membered with oxygen)
    "C1C(CO)C(CO)C(CO)C1",  # A polycyclic structure
    "CCO",  # Non-ring structure
    "C1CCCCCC1"  # Cycloheptane (7-membered)
]

valid_5_membered, valid_6_membered = filter_molecules(fullrecsmiles)

print(f"Valid molecules with 5-membered rings: {len(valid_5_membered)}")
print(valid_5_membered)

print(f"Valid molecules with 6-membered rings: {len(valid_6_membered)}")
print(valid_6_membered)
```

```python
from rdkit import Chem

def get_atom_counts(molecule):
    # Get the total number of atoms in the molecule
    num_atoms = molecule.GetNumAtoms()
    
    # Count the number of carbon atoms in the molecule
    num_carbons = sum(1 for atom in molecule.GetAtoms() if atom.GetSymbol() == 'C')
    
    return num_atoms, num_carbons

def calculate_averages(smiles_list):
    total_atoms = 0
    total_carbons = 0
    valid_molecule_count = 0
    
    for smiles in smiles_list:
        # Convert the SMILES string to a molecule object
        molecule = Chem.MolFromSmiles(smiles)
        
        if molecule is None:
            continue
        
        num_atoms, num_carbons = get_atom_counts(molecule)
        
        # Add to the total atom and carbon counts
        total_atoms += num_atoms
        total_carbons += num_carbons
        valid_molecule_count += 1
    
    if valid_molecule_count == 0:
        print("No valid molecules found.")
        return
    
    # Calculate averages
    avg_atoms = total_atoms / valid_molecule_count
    avg_carbons = total_carbons / valid_molecule_count
    
    print(f"Average total atom count: {avg_atoms:.2f}")
    print(f"Average carbon atom count: {avg_carbons:.2f}")



# Calculate and print the averages
calculate_averages(fullrecsmiles)
```

```python
def count_species_with_nitrogen(smiles_list):
    nitrogen_species_count = 0
    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)  # Try to create a molecule from the SMILES string
        if mol:  # If molecule is valid (even if partial)
            # Check if the molecule has any nitrogen atoms
            if any(atom.GetSymbol() == 'N' for atom in mol.GetAtoms()):
                nitrogen_species_count += 1
    return nitrogen_species_count

```

```python
def heteroatomsGeneral(list):
    heteroatomsCount = 0
    for i in range(len(list)):
        if 'P' in list[i]:
            heteroatomsCount+=1
            #print('S is there')
        elif 'N' in list[i]:
            heteroatomsCount+=1
        elif 'Si' in list[i]:
            heteroatomsCount+=1
        elif 'O' in list[i]:
            heteroatomsCount+=1
        elif 'S' in list[i]:
            heteroatomsCount+=1  
        elif 'si' in list[i]:
            heteroatomsCount+=1
        elif 'o' in list[i]:
            heteroatomsCount+=1
        elif 'p' in list[i]:
            heteroatomsCount+=1
        elif 'n' in list[i]:
            heteroatomsCount+=1

        continue   
    return print('there are ', heteroatomsCount, 'molecules with a heteroatomic substitution.')        
  #works and matches kelvin's 
```
